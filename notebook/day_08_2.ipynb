## Example

```python
# make the 2D map 

ant_map_input = '''
............
........0...
.....0......
.......0....
....0.......
......A.....
............
............
........A...
.........A..
............
............
''' 

ant_map = [ list(a) for a in ant_map_input.split() ]

ant_map 
```

```python
# how many different antennas (freq) are there

antennas_types = set(list(ant_map_input))
antennas_types = [a for a in antennas_types if a.isalnum()]
antennas_types
```

```python
# find antenna locations

ant_coords = dict() 

for ant_type in antennas_types:
    ant_coords[ant_type] = list() 
    
    for x in range(len(ant_map[0])):
        for y in range(len(ant_map)):
            if ant_map[y][x] == ant_type: ant_coords[ant_type].append( (x, y) )
print(ant_coords) 
```

```python
# Find antinodes for each pair or antennas 

antinodes = d8.find_antinodes_harmonics(ant_map, ant_coords, True)

        
```

```python
antinodes = set( [a2 for a1, a2, _, _ in antinodes] )
print(len(antinodes))
antinodes
```

```python
fp = '../data/day_08_1.txt'

with open(fp, 'r') as f_in:
   ant_map_input = f_in.readlines()

ant_map = [ list(a.strip()) for a in ant_map_input ]
#ant_map
```

```python
# how many different antennas (freq) are there

antennas_types = set(list(''.join(ant_map_input)))
antennas_types = [a for a in antennas_types if a.isalnum()]
print( antennas_types )
```

```python
# find antenna locations

ant_coords = dict() 

for ant_type in antennas_types:
    ant_coords[ant_type] = list() 
    
    for x in range(len(ant_map[0])):
        for y in range(len(ant_map)):
            if ant_map[y][x] == ant_type: ant_coords[ant_type].append( (x, y) )
print(ant_coords) 
```

```python
# Find antinodes for each pair or antennas 

antinodes = d8.find_antinodes_harmonics(ant_map, ant_coords)
```
