```python
# get the disk representation

disk_rep_input = '2333133121414131402'

disk_rep = [int(d) for d in list(disk_rep_input)]
disk_rep
```

```python
SPACE = '.'

def print_disk(disk_map):
    for id, d in disk_map:
        print(str(id)*d, end='')
    print('')
    
    
def get_disk_map(disk_rep):
    FLAG_FILE = True 
    result = list() 
    id = 0
    
    for d in disk_rep:
        if FLAG_FILE: 
            result.append( (id, d) )
            id = id + 1
        else:         
            if d>0: result.append( (SPACE, d) )
            
        FLAG_FILE = not FLAG_FILE
        
    return result
        
        
def is_block_continuous(disk_map):
    idx_space = disk_map.index(SPACE)
    
    for d in disk_map[idx_space+1:]:
        if d != SPACE: return False
        
    return True 


def find_space(disk_map, length): 
    try: 
        idx_start = disk_map.index(SPACE)
        
        while idx_start <= len(disk_map)-1: 
            spaces = disk_map[idx_start:idx_start+length]
            spaces = list(set(spaces))
            if idx_start+length <= len(disk_map)-1 and len(spaces) == 1 and spaces[0] == SPACE: return idx_start
            else: idx_start = disk_map.index(SPACE, idx_start+length)
    except ValueError:
        return -1
    return -1
    
disk_map = get_disk_map(disk_rep)
print_disk(disk_map )
```

```python
# find file IDs

file_ids = sorted([id for id, d in disk_map if id != SPACE], reverse=True) 
file_ids
```

```python
# defragement 

for f_id in file_ids:
    print_disk(disk_map)
    length = [(id, d) for id, d in disk_map if id == f_id][0][1]
    idx_f_id = disk_map.index( (f_id, length)  )
    
    for idx, segment in enumerate(disk_map): 
        if segment[0] == SPACE and segment[1] >= length and idx < idx_f_id:
            disk_map[idx] = (f_id, length)
            
            if segment[1] == length:
                disk_map[idx_f_id] = (SPACE, length)
                
            if segment[1] > length:
                disk_map.insert(idx+1, (SPACE, segment[1]-length) )
                disk_map[idx_f_id+1] = (SPACE, length)
                
            FLAG_CHANGED = True
            while FLAG_CHANGED:
                FLAG_CHANGED = False 
                for i in range(len(disk_map)-1): 
                    if disk_map[i][0] == SPACE and disk_map[i+1][0] == SPACE:
                        length_current = disk_map[i][1]
                        length_next = disk_map[i+1][1] 
                        disk_map.pop(i+1)
                        disk_map[i] = (SPACE, length_current+length_next)
                        FLAG_CHANGED = True
                        break 
                    
            break 
    
                
```

```python
# 2858 

idx = 0
result = 0 

for f_id, d in disk_map: 
    for i in (range(d)): 
        if f_id != SPACE: result = result + f_id * idx 
        idx = idx + 1 
result 
```
